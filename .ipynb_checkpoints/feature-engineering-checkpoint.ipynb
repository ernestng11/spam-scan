```python
import pandas as pd
import numpy as np

df = pd.read_csv('./spam.csv', encoding="latin-1")
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis=1, inplace=True)
df.columns = ['label', 'text']
```

```python
df['text_num_uppercase'] = df['text'].str.count(r'[A-Z]')
df['text_len'] = df['text'].str.len()
df['capital_pct'] = df.text_num_uppercase.div(df.text_len)

import seaborn as sns
import matplotlib.pyplot as plt
x1 = df.loc[df['label']=='spam']['capital_pct']
x2 = df.loc[df['label'] == 'ham']['capital_pct']

f, ax = plt.subplots(figsize=(10,5))
plot = sns.distplot(x1, label='Capital letters in Spam', color='r')
plot = sns.distplot(x2, label='Capital letters in Ham', color='g')
plt.legend()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,5))
plot = sns.violinplot(x='label',y='capital_pct', data=df)
plt.show()
```

```python
stop_words=set(open('./stopwords.txt','r').read().split('\n'))

from nltk.tag import pos_tag
from nltk import word_tokenize

df['text_num_stop_words'] = df['text'].str.split().apply(lambda x: len(set(x) & stop_words))
df['text_word_count'] = df['text'].apply(lambda x: len(str(x).split()))
df['stop_words_pct'] = df['text_num_stop_words'] / df['text_word_count']

x1 = df.loc[df['label']=='spam']['stop_words_pct']
x2 = df.loc[df['label'] == 'ham']['stop_words_pct']

f, ax = plt.subplots(figsize=(10,5))
plot = sns.distplot(x1, label='stopwords in Spam', color='r')
plot = sns.distplot(x2, label='stopwords in Ham', color='g')
plt.legend()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,5))
plot = sns.violinplot(x='label',y='stop_words_pct', data=df)
plt.show()
```

```python
#Part of Speech Counts

from textblob import TextBlob

pos_family = {
    'noun' : ['NN','NNS','NNP','NNPS'],
    'pron' : ['PRP','PRP$','WP','WP$'],
    'verb' : ['VB','VBD','VBG','VBN','VBP','VBZ'],
    'adj' :  ['JJ','JJR','JJS'],
    'adv' : ['RB','RBR','RBS','WRB']
}

# function to check and get the part of speech tag count of a words in a given sentence
def check_pos_tag(x, flag):
    cnt = 0
    try:
        wiki = TextBlob(x)
        for tup in wiki.tags:
            ppo = list(tup)[1]
            if ppo in pos_family[flag]:
                cnt += 1
    except:
        pass
    return cnt

df['noun_count'] = df['text'].apply(lambda x: check_pos_tag(x, 'noun'))
df['verb_count'] = df['text'].apply(lambda x: check_pos_tag(x, 'verb'))
df['adj_count'] = df['text'].apply(lambda x: check_pos_tag(x, 'adj'))
df['adv_count'] = df['text'].apply(lambda x: check_pos_tag(x, 'adv'))
df['pron_count'] = df['text'].apply(lambda x: check_pos_tag(x, 'pron'))
```

```python
postag_list=['noun_count', 'verb_count', 'adj_count', 'adv_count',
       'pron_count']
for cat in postag_list:
    print("Violin Plot for %s" % (cat))
    f, ax = plt.subplots(figsize=(10,5))
    plot = sns.violinplot(x='label', y=cat, data=df)
    plt.show()
```

Consider taking noun, verb and adj counts. adv and pronoun counts have approximately the same distribution for both spam and ham so we may not need to use those features.
